```python
import duckdb
import pandas as pd
import os
from datetime import datetime
```

```python
con = duckdb.connect(database = 'dados_duckdb.db', read_only = False)
```

```python
arquivo = 'z0019_2.csv'
data_ingestao = datetime.now()
df = pd.read_csv(f'../landing/{arquivo}', sep = ';')
df['nome_arquivo'] = arquivo
df['data_ingestao'] = data_ingestao
df.head()
```

```python
con.execute("""
create table if not exists bronze_produt(
            NATBR varchar,
            MAKTX varchar,
            WERKS varchar,
            MAINS varchar,
            LABST varchar,
            nome_arquivo varchar,
            data_ingestao timestamp)
""")
```

```python
con.execute("insert into bronze_produt select * from df")
```
